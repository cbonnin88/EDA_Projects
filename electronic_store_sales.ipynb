```python
import pandas as pd
import plotly.express as px
import gdown as gd
```

```python
url = 'https://drive.google.com/uc?id=1P_6a_9XnlsRRIfELCObdzdYz-G4Bh5ar'
gd.download(url,'electronics_store.csv',quiet=True)

df_electronic_sales = pd.read_csv('electronics_store.csv')
```

```python
# Converting 'Order Date' to a datetime
df_electronic_sales['Order Date'] = pd.to_datetime(df_electronic_sales['Order Date'], format='mixed', dayfirst=True)
df_electronic_sales.info()
```

```python
# Convert columns to string type before applying string methods
df_electronic_sales['Unit Cost'] = df_electronic_sales['Unit Cost'].astype(str).str.replace('€','',regex=False).str.replace(',','',regex=False)
df_electronic_sales['Price'] = df_electronic_sales['Price'].astype(str).str.replace('€','',regex=False).str.replace(',','',regex=False)

# Cleaning up the cost of sales, Sales, Profit column
df_electronic_sales['Cost of Sales'] = df_electronic_sales['Cost of Sales'].astype(str).str.replace('€','',regex=False).str.replace(',','',regex=False)
df_electronic_sales['Sales'] = df_electronic_sales['Sales'].astype(str).str.replace('€','',regex=False).str.replace(',','',regex=False)
df_electronic_sales['Profit'] = df_electronic_sales['Profit'].astype(str).str.replace('€','',regex=False).str.replace(',','',regex=False)

df_electronic_sales.head()
```

```python
df_electronic_sales.info()
```

```python
# Converting certain columns to floats
df_electronic_sales['Unit Cost']=df_electronic_sales['Unit Cost'].astype(float)
df_electronic_sales['Price']=df_electronic_sales['Price'].astype(float)

# Cleaning up the cost of sales, Sales, Profit column
df_electronic_sales['Cost of Sales'] = df_electronic_sales['Cost of Sales'].str.replace('[^0-9\.]', '', regex=True)
df_electronic_sales['Sales'] = df_electronic_sales['Sales'].str.replace('[^0-9\.]', '', regex=True)
df_electronic_sales['Profit'] = df_electronic_sales['Profit'].str.replace('[^0-9\.]', '', regex=True)


df_electronic_sales['Cost of Sales']=df_electronic_sales['Cost of Sales'].astype(float)
df_electronic_sales['Sales']= df_electronic_sales['Sales'].astype(float)
df_electronic_sales['Profit']=df_electronic_sales['Profit'].astype(float)

df_electronic_sales.info()
```

```python
df_electronic_sales.head()
```

```python
df_electronic_sales.dtypes
```

```python
# Converting to Strings
df_electronic_sales['Channel']=df_electronic_sales['Channel'].astype('string')
df_electronic_sales['Promotion Name'] = df_electronic_sales['Promotion Name'].astype('string')
df_electronic_sales['Profit'] = df_electronic_sales['Profit'].astype('string')
df_electronic_sales['Manufacturer'] = df_electronic_sales['Manufacturer'].astype('string')
df_electronic_sales['Product Sub Category'] = df_electronic_sales['Product Sub Category'].astype('string')
df_electronic_sales['Product Category'] = df_electronic_sales['Product Category'].astype('string')
df_electronic_sales['Region'] = df_electronic_sales['Region'].astype('string')
df_electronic_sales['City'] = df_electronic_sales['City'].astype('string')
df_electronic_sales['Country']=df_electronic_sales['Country'].astype('string')


df_electronic_sales.info()
```

# **EDA and Data Visualizations**

```python
df_electronic_sales.nunique()
```

```python
# Histogram for the order years
df_electronic_sales['Year'] = df_electronic_sales['order_date'].dt.year
df_electronic_sales.info()
```

```python
fig_years = px.histogram(
    df_electronic_sales,
    x='Year',
    title='Distribution of Sales by Year',
    color_discrete_sequence=['#40826D']
)

fig_years.update_xaxes(dtick=1)

fig_years.show()
```

```python
product_counts = df_electronic_sales['product_category'].value_counts()
```

```python
# The number of items in each 'product_category'
fig_category = px.bar(
    product_counts,
    y='count',
    title='Number of Products in Each Category',
    labels={'product_category':'Products','count':'Number of Products'},
    color='count',
    color_continuous_scale='Viridis',
    text_auto=True
)

fig_category.update_layout(coloraxis_showscale=False)

fig_category.show()
```

Calculating the average profit by Region

```python
display(df_electronic_sales.columns)
```

```python
df_electronic_sales['profit'] = pd.to_numeric(df_electronic_sales['profit'], errors='coerce')

average_profit_by_region = df_electronic_sales.groupby('region')["profit"].mean().round(0).sort_values(ascending=False)
display(average_profit_by_region)
```

```python
fig_profit_region_box = px.box(
    df_electronic_sales,
    x='region',
    y='profit',
    title='Distribution of Profit by Region',
    labels={'region':'Region', 'profit':'Profit'},
    color='region',
    color_discrete_sequence=[ '#84a625','#d36b42','#1e97ba']
)

fig_profit_region_box.show()
```

```python
fig_price_sales = px.scatter(
    df_electronic_sales,
    x='price',
    y='sales',
    title='Sales vs Price',
    labels={'price':'Price', 'sales':'Sales'},
    hover_data=['product_name', 'region', 'channel'],
    color='price'
)

fig_price_sales.update_layout(coloraxis_showscale=False)

fig_price_sales.show()
```

```python
country_count = df_electronic_sales['country'].value_counts().sort_values(ascending=False).head(5)
display(country_count)
```

```python
country_count_df = country_count.reset_index()
country_count_df.columns = ['country','count']

fig_country = px.pie(
    country_count_df,
    names='country',
    values='count',
    title='Top 5 Countries by Sales',

)

fig_country.update_traces(hovertemplate='Country: %{label}<br>Number of Sales: %{value}',
                          textinfo='label+percent',textfont_size=12)

fig_country.show()
```

```python
europe_sales = df_electronic_sales[df_electronic_sales['region'] == 'Europe']['country'].value_counts().sort_values(ascending=False).head(5)
display(europe_sales)
```

```python
europe_count_df = europe_sales.reset_index()
europe_count_df.columns = ['country','count']

fig_country = px.pie(
    europe_count_df,
    names='country',
    values='count',
    title='Top 5 European Countries by Sales',

)

fig_country.update_traces(hovertemplate='Country: %{label}<br>Number of Sales: %{value}',
                          textinfo='label+percent',textfont_size=12)

fig_country.show()
```

```python
df_electronic_sales['Year'] = df_electronic_sales['Year'].astype(str)

sale_over_time = df_electronic_sales.groupby('Year')['sales'].sum()
display(sale_over_time)
```

```python
sales_over_time_df = sale_over_time.reset_index()
sales_over_time_df.columns = ['Year','sales']

fig_time = px.line(
    sales_over_time_df,
    x='Year',
    y='sales',
    markers=True

)

fig_time.show()
```
